```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
# Para que aparezcan en el notebook
%matplotlib inline 
#from itertools import repeat
#import math
```

El sistema de ecuaciones diferenciales del quimiostato está dado por:
    


$\frac{d S}{dt}= (S_{0} - S) \, D - \frac{\mu_1}{y_{1}} \, \frac{S \cdot N_{1}}{K_{S_{1}}+S} - \frac{\mu_2}{y_{2}} \, \frac{S \cdot N_{2}}{K_{S_{2}}+S} \\ $

$\frac{dN_{1}}{dt} =  \frac{\mu_1 S \cdot N_{1}}{K_{S_{1}}+S} - D \cdot N_{1}$

$\frac{dN_{2}}{dt} =  \frac{\mu_2 S \cdot N_{2}}{K_{S_{2}}+S} - D \cdot N_{2}$


## Un solo recurso y una sola bacteria

```python

# Resolver sistema de ecuaciones diferenciales

def Change(x, t, params):
    
    bacteria, substrate = x  
    dSdt = params['d']*(params['S']-substrate) - (1/c * params['Vmax'] * substrate/(params['Km'] + substrate) * bacteria)
    dBdt = params['Vmax'] * substrate/(params['Km'] + substrate)*bacteria - params['d']*bacteria
    
    return [dBdt, dSdt]


# Gráficas
def plot1_1 (xt, nexp):
    plt.subplot(121)
    plt.xlabel("Time")
    plt.ylabel("Y")
    plt.plot(t, [row[0] for row in xt], label='Bacteria', color='c')
    plt.legend(loc='upper right')
    plt.title("Bacteria in "+nexp)
    
    plt.subplot(122)
    plt.xlabel("Time")
    plt.ylabel("Y")
    plt.plot(t, [row[1] for row in xt], label='Substrate', color='m')
    plt.legend(loc='upper right')
    plt.title("Substrate in "+nexp)
    
    plt.show()
    
    plt.xlabel("Bacteria")
    plt.ylabel("Substrate")
    plt.plot([row[0] for row in xt],[row[1] for row in xt],label="Population", color='m');
    plt.legend(loc='upper right')
    plt.title('Phase plane')
```

### Para experimento 1: C-8


```python
T=100   # Tiempo
nT=100 # Número de pasos de integración
S=1e-4
d=6e-2
Vmax=0.81
c=2.5e10 # c=yield
Km=3e-6

#Condiciones iniciales
B=1e6
So=0
x0 = [B, So]    

```

```python
params={'S': S, 'd':d, 'Vmax':Vmax, 'c':c, 'Km':Km}
t = np.linspace(0,T,nT) # Crea un vec de 0 a T y te manda en cuantas partes lo hace    
xt = odeint(Change, x0, t, args=(params,))
plot1_1(xt, "C-8")
```

## Para dos bacterias compitiendo por un recurso

```python
# Resolver sistema de ecuaciones diferenciales. 
def Change2(x, t, params1):
    
    bacteria1, bacteria2, substrate = x
    
    dB1dt = ((params['Vmax1'] * substrate*bacteria1)/(params['Km1'] + substrate)) - params['d']*bacteria1
    dB2dt = ((params['Vmax2'] * substrate*bacteria2)/(params['Km2'] + substrate)) - params['d']*bacteria2
    dSdt = (params['d']*(params['S']-substrate)) - (params['Vmax1']/params['c1'])*((substrate*bacteria1)/params['Km1']+substrate) - (params['Vmax2']/params['c2'])*((substrate*bacteria2)/params['Km2']+substrate)

    return [dB1dt, dB2dt, dSdt]

# Graficar

def plot2_1(xt, nexp,b1, b2):
    np.seterr(divide = 'ignore') 


    plt.subplot(121)
    plt.xlabel("Time")
    plt.ylabel("Density")
    plt.title(nexp)
    plt.plot(t, [row[0] for row in xt], label=b1, color='c')
    plt.plot(t, [row[1] for row in xt], label=b2, color='b')
    plt.legend(loc='upper right')
    

    plt.subplot(122)
    plt.xlabel("Time")
    plt.ylabel("Density")
    plt.title(nexp + " (Log scale)")
    plt.plot(t, [row[0] for row in np.log10(xt)], label=b1, color='c')
    plt.plot(t, [row[1] for row in np.log10(xt)], label=b2, color='b')
    #plt.yscale('log')
    plt.legend(loc='upper right')

    plt.show()
```

Al simular numéricamente los tres experimentos, el parámetro que parece más determinante para que una cepa se imponga sobre otra parece ser $Vmax$ -la tasa máxima de generación de nuevas bacterias-, parecería natural que aquella con los mayores valores para este parámetro es la que logra sobrevivir.

Sin embargo, en el primer experimento, la cepa que prevalece es C-8, con un valor de $Vmax$ menor (0.81) que la cepa PAO283 (0.91). No obstante, esto cobra sentido al observar que su valor de $K_{m}$ -la constante de la mitad de la saturación- varía por casi dos órdenes de magnitud (3e-6 vs. 3.1e-4), lo cual refleja que tiene una mayor afinidad por el sustrato presente en el medio.
